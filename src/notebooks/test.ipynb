```python
import pandas as pd

df = pd.read_parquet("C:/Users/kenji/dev/sdm_bio_vinaceous-breasted-amazon/data/30-05-2024/features.parquet")
```

```python
def correlation(dataset: pd.DataFrame, threshold: float = 0.7):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr

threshold = 0.7

y = df.pop("target")
X = df.copy().drop(columns=['lat', 'lon'])

removed_columns = correlation(X, threshold= threshold)
X_CP = X.drop(columns=removed_columns)
```
