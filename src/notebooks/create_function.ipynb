```python
def treat_dataframe(dataframe_path: str,rename_columns: dict = {}):
    if rename_columns:
        data = pd.read_parquet(dataframe_path)[list(rename_columns.keys())].rename(columns=rename_columns)
    else:
        data = pd.read_parquet(dataframe_path)[['lat','lon']]
    data['target'] = 1
    return data
```
